```python
# Import the necessary libraries
import json
import numpy as np 
import os
import pandas as pd
import pyreadr 

# Import HINTS data and convert to Python
result = pyreadr.read_r("C:/Users/kyram/OneDrive/School/hints6_public.rda")

# Define DataFrame
df1 = result['public']
```

```python
# Condense sparse features
df1['AsAm'] = df1[['AsInd', 'Chinese', 'Filipino', 'Vietnamese', 'OthAsian']].isin(['Selected']).sum(axis=1) > 0

df1['Hisp'] = df1[['Mexican', 'PuertoRican', 'Cuban', 'OthHisp']].isin(['Selected']).sum(axis=1) > 0

df1['Sunburned_Work'] = df1[['Sunburned_JobOutside', 'Sunburned_HomeOutside']].isin(['Selected']).sum(axis=1) > 0

df1['Sunburned_Active'] = df1[['Sunburned_Exercise', 'Sunburned_Swimming']].isin(['Selected']).sum(axis=1) > 0

df1['Sunburned_Rec'] = df1[['Sunburned_Sunbathing', 'Sunburned_SportingEvent', 'Sunburned_OutdoorEvent']].isin(['Selected']).sum(axis=1) > 0

df1['Sunburned_Other'] = df1[['Sunburned_Other', 'Sunburned_DayToDay', 'Sunburned_DK']].isin(['Selected']).sum(axis=1) > 0
```

```python
# Define new DataFrame
df = df1[['Age', 'AsAm', 'AmerInd', 'AvgDrinksPerWeek', 'BirthGender', 'Black','DelayNeededCare', 'DocTalkLDCT', 'DocTellColorectalTests', 'DrinkDaysPerWeek', 'DrinksOneOccasion', 'DrinksPerDay', 'Education', 'EverHadCancer', 'FamilyEverHadCancer2', 'FreqGoProvider', 'HealthInsurance2', 'HeardHPVVaccine2', 'HPVCauseCancer_Cervical', 'Hisp', 'IncomeRanges', 'MedConditions_Diabetes', 'MedConditions_HighBP', 'MedConditions_HeartCondition', 'MedConditions_LungDisease', 'OthPacIsl', 'QualityCare', 'Smoke100', 'SmokeNow', 'Sunburned_Active', 'Sunburned_Other', 'Sunburned_Rec', 'Sunburned_Work', 'SunEffectAfter1Hour','TimesSunburned', 'UseECigNow', 'UsedECigEver', 'WhenPapTest', 'White']]
```

```python
# Get rid of null values
for col_name in df.columns:
    df.loc[df[col_name].isin(['Missing data (Web partial - Question Never Seen)', 'Missing data (Not Ascertained)', 'Missing data (Filter Missing)', 'Question answered in error (Commission Error)', 'Inapplicable, coded 2 in HeardHPV', 'Inapplicable, coded 1 in BirthGender', 'I am male (Web only)', 'Inapplicable, coded 2 in UsedECigEver', 'Unreadable or Nonconforming Numeric Response', 'Inapplicable, coded 0 in TimesSunburned', 'Inapplicable, coded 0 in FreqGoProvider', 'Inapplicable, coded 0 in DrinkDaysPerWeek', 'Inapplicable, coded 1 in HeardGenTest_None or coded 1 in Tes', 'Not selected', 'Not sure', 'Multiple responses selected in error']), col_name] = np.nan
    
else:
    print(f"Column '{col_name}' not found in DataFrame")
```

```python
# Convert strings to numbers
df['AvgDrinksPerWeek'] = df['AvgDrinksPerWeek'].replace('None', 0)

df['DrinksPerDay'] = df['DrinksPerDay'].replace('None', 0)

df['DrinkDaysPerWeek'] = df['DrinkDaysPerWeek'].replace('None', 0)

df['DrinksOneOccasion'] = df['DrinksOneOccasion'].replace('Never', 0)
df['DrinksOneOccasion'] = df['DrinksOneOccasion'].replace('1 or 2 times', 1.5)
df['DrinksOneOccasion'] = df['DrinksOneOccasion'].replace('3 to 5 times', 4)
df['DrinksOneOccasion'] = df['DrinksOneOccasion'].replace('6 to 10 times', 8)
df['DrinksOneOccasion'] = df['DrinksOneOccasion'].replace('11 or more times', 11)

df['Education'] = df['Education'].replace('Less than 8 years', 1)
df['Education'] = df['Education'].replace('8 through 11 years', 2)
df['Education'] = df['Education'].replace('12 years or completed high school', 3)
df['Education'] = df['Education'].replace('Post high school training other than college (vocational) or', 4)
df['Education'] = df['Education'].replace('Some college', 5)
df['Education'] = df['Education'].replace('College graduate', 6)
df['Education'] = df['Education'].replace('Postgraduate', 7)

df['FreqGoProvider'] = df['FreqGoProvider'].replace('None', 0)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('1 time', 1)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('2 times', 2)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('3 times', 3)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('4 times', 4)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('5-9 times', 5)
df['FreqGoProvider'] = df['FreqGoProvider'].replace('10 or more times', 6)

df['IncomeRanges'] = df['IncomeRanges'].replace("$0 to $9,999", 4999)
df['IncomeRanges'] = df['IncomeRanges'].replace('$10,000 to $14,999', 12499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$15,000 to $19,999', 17499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$20,000 to $34,999', 27499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$35,000 to $49,999', 42499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$50,000 to $74,999', 62499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$75,000 to $99,999', 87499)
df['IncomeRanges'] = df['IncomeRanges'].replace('$100,000 to $199,999', 149999)
df['IncomeRanges'] = df['IncomeRanges'].replace('$200,000 or more', 200000)

df['WhenPapTest'] = df['WhenPapTest'].replace("A year ago or less", 1)
df['WhenPapTest'] = df['WhenPapTest'].replace('More than 1, up to 2 years ago', 2)
df['WhenPapTest'] = df['WhenPapTest'].replace('More than 2, up to 3 years ago', 3)
df['WhenPapTest'] = df['WhenPapTest'].replace('More than 3, up to 5 years ago', 4)
df['WhenPapTest'] = df['WhenPapTest'].replace('More than 5 years ago', 6)
df['WhenPapTest'] = df['WhenPapTest'].replace('I have never had a Pap test', 20)

df['TimesSunburned'] = df['TimesSunburned'].replace('No sunburns in past 12 months', 0)
```

```python
# Convert Booleans to numbers
columns_yn = ['EverHadCancer', 'FamilyEverHadCancer2','FreqGoProvider', 'HealthInsurance2', 'HeardHPVVaccine2', 'HPVCauseCancer_Cervical', 'MedConditions_Diabetes', 'MedConditions_HeartCondition', 'MedConditions_HighBP', 'MedConditions_LungDisease', 'Smoke100',  'UsedECigEver']

columns_tf = ['Sunburned_Active', 'Sunburned_Other', 'Sunburned_Rec', 'Sunburned_Work']

columns_sel = ['AsAm', 'AmerInd', 'Black', 'Hisp', 'OthPacIsl', 'White']
for col_name in columns_yn:

    df[col_name] = df[col_name].replace('No', 0)

    df[col_name] = df[col_name].replace('Yes', 1)
    
    df[col_name] = pd.to_numeric(df[col_name], errors= 'coerce')
    
    df[col_name] = df[col_name].fillna(0)


for col_name in columns_tf:

    df[col_name] = df[col_name].replace(False, 0)

    df[col_name] = df[col_name].replace(True, 1)
    
    df[col_name] = pd.to_numeric(df[col_name], errors= 'coerce')
    
    df[col_name] = df[col_name].fillna(0)

for col_name in columns_sel:
    df[col_name] = df[col_name].replace('Selected', 1)
        
    df[col_name] = pd.to_numeric(df[col_name], errors= 'coerce')
    
    df[col_name] = df[col_name].fillna(0)

```

```python
# Convert strings to numbers
numeric = ['Age', 'AvgDrinksPerWeek', 'DrinksOneOccasion', 'DrinksPerDay', 'DrinkDaysPerWeek', 'Education', 'IncomeRanges','TimesSunburned', 'WhenPapTest']

for col_name in numeric:
    df[col_name] = pd.to_numeric(df[col_name], errors= 'coerce')
    
    df[col_name] = df[col_name].fillna(df[col_name].median())
```

```python
# List the columns that need to be encoded
columns_to_encode = ['BirthGender', 'DelayNeededCare', 'DocTalkLDCT', 'DocTellColorectalTests', 'QualityCare', 'SmokeNow', 'SunEffectAfter1Hour', 'UseECigNow']
```

```python
# OneHotEncode columns
for col_name in columns_to_encode:
    df[col_name] = df[col_name].cat.add_categories('Missing')

    df[col_name] = df[col_name].fillna('Missing')

    df[col_name] = df[col_name].cat.remove_unused_categories()

    df[col_name] = df[col_name].cat.remove_categories('Missing')
        
    df_encoded = pd.get_dummies(df[col_name],prefix =col_name, prefix_sep='_')
    
    df = pd.concat([df, df_encoded], axis=1
                   )
    df.drop(col_name, axis=1, inplace=True)
```

```python
# Convert DataFrame to a dictionary
preprocessed_data = {
    'public': df.to_dict(),  # Converting DataFrame to a dictionary
}
# Define the save path
save_path = r"C:\Users\kyram\OneDrive\School\Prepossesed Data.json"

# Ensure the directory exists
os.makedirs(os.path.dirname(save_path), exist_ok=True)

# Save the preprocessed data to the JSON file
with open(save_path, 'w') as file:
    json.dump(preprocessed_data, file)

print(f"Data successfully saved to {save_path}")
```
