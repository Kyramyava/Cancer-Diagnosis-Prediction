```python
import json
import os
import pandas as pd

# Define the file path
file_path = r"C:\Users\kyram\OneDrive\School\RF_importance.json"

# Confirm the file exists
if not os.path.exists(file_path):
    print(f"The file does not exist at the specified path: {file_path}")
else:
    # Load the JSON file
    with open(file_path, 'r') as file:
        RF_importance = json.load(file)

    # Print the keys 
    print(RF_importance.keys())

    # Access the keys
    if 'public' in RF_importance:
        public_data = RF_importance['public']
        # Convert the dictionary back to a DataFrame
        df_RF1 = pd.DataFrame.from_dict(public_data)
        print("df successful")
    else:
        print("'public' key not found in the JSON file")
```

```python
# Define the path
file_path = r"C:\Users\kyram\OneDrive\School\FNN_importance.json"

# Confirm the file exists
if not os.path.exists(file_path):
    print(f"The file does not exist at the specified path: {file_path}")
else:
    # Load the JSON file
    with open(file_path, 'r') as file:
        FNN_importance = json.load(file)

    # Print the keys 
    print(FNN_importance.keys())

    # Access the keys
    if 'public' in FNN_importance:
        public_data = FNN_importance['public']
        # Convert the dictionary back to a DataFrame
        df_FNN1 = pd.DataFrame.from_dict(public_data)
        print("df successful")
    else:
        print("'public' key not found in the JSON file")
```

```python
# Import necessary library
import matplotlib.pyplot as plt
```

```python
# Define the DataFrames
df_RF1 = df_RF1.sort_values(by='importance', ascending=True)

df_FNN1 = df_FNN1.sort_values(by='importance', ascending=True)
```

```python
# Define features by their rank and drop the specific importance value
df_RF1['RF 1 Rank'] = df_RF1['importance'].rank(ascending=False)
df_FNN1['FNN 2 Rank'] = df_FNN1['importance'].rank(ascending=False)
df_RF1_ranked = df_RF1.drop(columns=['importance'])
df_FNN1_ranked = df_FNN1.drop(columns=['importance'])

# Merge the DataFrames on the feature name
merged_df1 = pd.merge(df_RF1_ranked, df_FNN1_ranked, on='feature').sort_values(by= 'feature')

# Display the merged DataFrame
print(merged_df1)
```

```python
# Display feature importance as a graph
plt.figure(figsize=(5, 20))
plt.plot(df_RF1['importance'], df_RF1['feature'], 'o-', label='Importance')
plt.ylabel('Feature')
plt.xlabel('Importance')
plt.title('RF 1 Feature Importance Rankings')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Display feature importance as a graph
plt.figure(figsize=(5, 20))
plt.plot(df_FNN1['importance'], df_FNN1['feature'], '>-', label='Importance')
plt.ylabel('Feature')
plt.xlabel('Importance')
plt.title('FNN 1 Importance Rankings')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
